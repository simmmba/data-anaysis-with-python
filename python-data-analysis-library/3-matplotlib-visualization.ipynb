# matplotlib : visualization (시각화 라이브러리)

- 데이터를 시각화할 때 사용하는 라이브러리
- 곡선, 원, 막대그래프 등

```python
import matplotlib.pyplot as plt
```

## 간단한 그래프

```python
x = [1, 2, 3]
y = [0, 1, 2]

plt.plot(x, y) # x축, y축 그래프 그리기
plt.show() # 그래프 보여주기
```

### sin 곡선

```python
import numpy as np
import matplotlib.pyplot as plt

# np.arange(시작점, 끝점, 증가량) : 리스트 만들 때
x = np.arange(0, 10, 0.1)
print(x)

# np.sin(값) : sin값
y = np.sin(x)
print(y)

plt.plot(x, y)
plt.show()
```

### 여러 그래프 같이 그리기

```python
t = np.arange(0., 10., 0.2)

# 여러 점 동시에 그리기. 
plt.plot(t, t, 'r--', # x축 : t, y축 : y 
         t, 2 * t, 'bs',  # x축 : t, y축 : 2 * t
         t, t**2, 'g^') # x 축 : t, y 축 : t^2

plt.title('several graphs') # 그래프에 제목 붙이기          
plt.show()
```

## 막대 그래프

```python
import numpy as np
import pandas as pd
import random
import matplotlib.pyplot as plt

# 랜덤인 정수(정수 범위 0~100)가 50개 있는 리스트 만들기 
age = [random.randint(20, 100) for x in range(0, 50)]
weight = [random.randint(40, 100) for i in range(0, 50)]   

df = pd.DataFrame({
    'age': age,
    'weight': weight
})

# pandas 라이브러리 써서 그래프 그리기
df.plot(kind = 'bar', subplots = True) # subplots : 그래프 둘로 나눠서 보여주기
plt.title('Bar Chart') # 차트에 제목 붙이기
plt.savefig('bar_chart_random.png') # 차트 이미지로 저장
plt.show() 
```
